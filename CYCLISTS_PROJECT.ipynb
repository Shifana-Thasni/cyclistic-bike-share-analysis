```python
import pandas as pd
import numpy as num
```

### Renaming Columns
The column names in the 2019 dataset are renamed to match the 2020 dataset structure for consistency.

```python
df_2019=df_2019.rename(columns={ 'trip_id':'ride_id',
                                'bike_id':'ridable_type',
                               'start_time':'started_at',
                               'end_time':'ended_at',
                               'usertype':'member_casual'})
```

### Describe the data

```python
df_2019.describe()
```

### Checking Missing Values
The dataset is checked for null or missing values in each column.

```python
print(df_2020.isnull().sum())
```

### Merging Datasets
The 2019 and 2020 datasets are combined into a single dataset for analysis.

```python
all_trips=pd.concat([df_2019,df_2020],ignore_index=True)
```

### Previewing the Dataset
The first five rows of the merged dataset are displayed for inspection.

```python
all_trips.head(5)
```

### Calculating Ride Duration
A new column named ride_length is created by calculating the difference between trip end time and start time.

```python
all_trips['ride_length']=pd.to_datetime(all_trips['ended_at'])-pd.to_datetime(all_trips['started_at'])
```

### Viewing Ride Length
The first few values of the ride_length column are displayed for verification.

```python
print(all_trips['ride_length'].head())
```

### Day of Week Extraction
Converting 'started_at' to datetime and extracting 
the day of the week for ride pattern analysis.

```python
all_trips['day_of_week']=pd.to_datetime(all_trips['started_at']).dt.dayofweek
print(all_trips['day_of_week'].head())
```

### Filtering Invalid Rides
Removing rides with zero or negative duration, 
keeping only trips lasting at least 1 minute.

```python
all_trips=all_trips[all_trips['ride_length']>pd.Timedelta(0)]
all_trips=all_trips[all_trips['ride_length']>= pd.Timedelta(minutes=1)]
print(all_trips.isnull().sum())
                    
```

### Checking Duplicates
Verifying there are no duplicate rows in the dataset.

```python
all_trips.duplicated().sum()
```

### Selecting Relevant Columns
Keeping only the 6 columns needed for analysis.

```python
cols_to_keep=['ride_id','started_at','ended_at','member_casual','ride_length','day_of_week']
all_trips=all_trips[['ride_id','started_at','ended_at','member_casual','ride_length','day_of_week']]
```

### Final Dataset Shape
Confirming the cleaned dataset has 784,285 rows and 6 columns.

```python
print("final shape:" ,all_trips.shape)
```

### Summary Statistics
Reviewing ride length distribution, most common 
day, and member type breakdown.

```python
print(all_trips['ride_length'].describe())
print("mode:",all_trips['day_of_week'].mode()[0])
print(all_trips['member_casual'].value_counts())
```

### Converting Ride Length to Minutes
Converting ride length from timedelta to 
numeric minutes for easier analysis.

```python
all_trips['ride_length_minutes']=all_trips['ride_length'].dt.total_seconds()/60
```

### Ride Patterns Analysis
Calculating average ride length by member type 
and total rides by day of week.

```python
print("=== AVERAGE RIDE LENGTH BY MEMBER TYPE===")
print(all_trips.groupby('member_casual')['ride_length_minutes'].mean())
print("\n===RIDES BY DAY OFWEEK===")
print(all_trips['day_of_week'].value_counts().sort_index())
```

### Standardizing Member Types
Renaming 'Subscriber' to 'member' and 
'Customer' to 'casual' for consistency.

```python
all_trips['member_casual']=all_trips['member_casual'].replace({
    'Subscriber':'member',
    'Customer':'casual'
})

print("member typesafter fix:")
print(all_trips['member_casual'].value_counts())
```

### Adding Ride Count Column
Creating a count column with value 1 
for each ride record.

```python
all_trips['count_ride_id]']=1
```

### Extracting Month
Converting 'started_at' to datetime and 
extracting the month for trend analysis.

```python
all_trips['started_at'] = pd.to_datetime(all_trips['started_at'])
all_trips['month'] = all_trips['started_at'].dt.month
print(all_trips['month'].value_counts().sort_index())
```

### Checking Shape
Verifying the final dimensions of the dataset.

```python
print(all_trips.shape)
```

### Average Ride Length by Member Type

```python
print(all_trips.groupby('member_casual')['ride_length_minutes'].mean())
```

### Extracting Month Number
Pulling the month from 'started_at' 
for monthly trend analysis.

```python
all_trips['month'] = all_trips['started_at'].dt.month
```

### Creating Summary Table
Grouping rides by month, day of week, 
and member type, then exporting to CSV.

```python
summary = all_trips.groupby(
    ['month', 'day_of_week', 'member_casual']
)['ride_id'].count().reset_index()

summary.columns = ['month', 'day_of_week', 
                   'member_casual', 'count_ride_id']

summary.to_csv("final_cyclistic1_summary.csv", index=False)
print("Exported successfully!")
print(summary.shape)
```
